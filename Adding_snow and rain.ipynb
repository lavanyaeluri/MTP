```python
import cv2
import numpy as np

#Adding snow   
def snow(frame):    
    frame_HLS = cv2.cvtColor(frame,cv2.COLOR_RGB2HLS)
    cv2.imshow("FRAME_HLS", frame_HLS)
    frame_HLS = np.array(frame_HLS, dtype = np.float64)     
    brightness_coefficient = 2.5     
    snow_point=100     
    frame_HLS[:,:,1][frame_HLS[:,:,1]<snow_point] = frame_HLS[:,:,1][frame_HLS[:,:,1]<snow_point]*brightness_coefficient  
    frame_HLS[:,:,1][frame_HLS[:,:,1]>255]  = 255   
    frame_HLS = np.array(frame_HLS, dtype = np.uint8)   
    frame_RGB = cv2.cvtColor(frame_HLS,cv2.COLOR_HLS2RGB)    
    return frame_RGB

#Adding rain
def random_lines(imshape,slant,drop_length):   
    drops=[]    
    for i in range(1500):        
        if slant<0:            
            x= np.random.randint(slant,imshape[1])        
        else:            
            x= np.random.randint(0,imshape[1]-slant)        
            y= np.random.randint(0,imshape[0]-drop_length)       
            drops.append((x,y))    
        return drops            
def rain(frame): 
                global frame_RGB
                imshape = frame.shape   
                slant_extreme=5    
                slant= np.random.randint(-slant_extreme,slant_extreme)    
                drop_length=15   
                drop_width=2   
                drop_color=(200,200,200)    
                rain_drops=random_lines(imshape,slant,drop_length)       
                for rain_drop in rain_drops:       
                    cv2.line(frame,(rain_drop[0],rain_drop[1]),(rain_drop[0]+slant,rain_drop[1]+drop_length),drop_color,drop_width)    
                    frame= cv2.blur(frame,(7,7))         
                    brightness_coefficient = 0.7      
                    frame_HLS = cv2.cvtColor(frame,cv2.COLOR_RGB2HLS)    
                    frame_HLS[:,:,1] = frame_HLS[:,:,1]*brightness_coefficient     
                    frame_RGB = cv2.cvtColor(frame_HLS,cv2.COLOR_HLS2RGB) 
                return frame_RGB
            
cap = cv2.VideoCapture("C:/Users/Lavanya/Downloads/M.TECH Project/videos/video2.mp4")
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        #Resize the frame with width 640 and height 360
        #frame = cv2.resize(frame, (640, 360), interpolation = cv2.INTER_LINEAR)
        snow_frame=snow(frame)
        rain_frame=rain(frame)
        cv2.imshow("After", rain_frame)
        cv2.imshow("Before", frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else: break

cap.release()
cv2.destroyAllWindows()
```
